```python
# Group data by batsman and sum their runs
batsman_runs = df_icc.groupby('Batsman Name')['Runs'].sum().sort_values(ascending=False)

# Select the top 10 highest scorers
top_10_scorers = batsman_runs.head(10)

# Convert to DataFrame for Plotly
top_10_df = top_10_scorers.reset_index()
top_10_df.columns = ['Batsman', 'Runs']

# Create a bar chart for the top 10 highest run-scorers
plt.figure(figsize=(12, 8))
barplot = sns.barplot(x=top_10_scorers.index, y=top_10_scorers.values)  # Use default palette
plt.title('Top 10 Batsmen with Highest Run Scorers')
plt.xlabel('Batsman Name')
plt.ylabel('Total Runs Scored')
plt.xticks(rotation=45)

# Define a color palette
colors = sns.color_palette('viridis', n_colors=10)  # Get 10 colors from the 'viridis' palette

# Apply the colors to the bars and annotate the bars with total runs
for i, (bar, runs) in enumerate(zip(barplot.patches, top_10_scorers)):
    bar.set_color(colors[i])  # Set the color of each bar based on the viridis palette
    plt.text(bar.get_x() + bar.get_width() / 2,  # x-coordinate of text
             bar.get_height() + 0.5,             # y-coordinate of text (above the bar)
             f'{runs}',                          # Text to display (total runs)
             ha='center',                         # Horizontally centered
             va='bottom',                         # Vertically above the bar
             fontsize=10,                         # Font size
             color='black')                       # Text color

plt.show()
```

```python
# Step 1: Aggregate the count of wins and losses for each team
wins = df_icc[df_icc['Winner'] == df_icc['Team Innings']].groupby('Team Innings')['Winner'].count().sort_values(ascending=False)
losses = df_icc[df_icc['Winner'] != df_icc['Team Innings']].groupby('Team Innings')['Winner'].count().sort_values(ascending=False)

# Step 2: Select the top 5 teams with the highest counts for wins and losses
top_5_wins = wins.head(5)
top_5_losses = losses.head(5)

# Step 3: Visualize the top 5 teams with bar plots
plt.figure(figsize=(10, 5))

# Bar plot for top 10 teams with the most wins
plt.subplot(1, 2, 1)
top_5_wins.plot(kind='bar', color='green')
plt.title('Top 5 Teams with the Most Wins')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)

# Bar plot for top 5 teams with the most losses
plt.subplot(1, 2, 2)
top_5_losses.plot(kind='bar', color='red')
plt.title('Top 5 Teams with the Most Losses')
plt.xlabel('Team')
plt.ylabel('Number of Losses')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Creating a new DataFrame for the ML modeling
ml_data = df_icc[['Batting Position', 'Balls', '4s', '6s', 'Strike Rate', 'Ground', 'Runs']].copy()

# Encoding categorical variables
label_encoder = LabelEncoder()
ml_data['Ground'] = label_encoder.fit_transform(ml_data['Ground'])

# Check for any missing values
missing_data = ml_data.isnull().sum()

# Display the encoded data and any missing values
ml_data.head(), missing_data
```

```python
# Define bins for the 'runs' categories
bins = [-1, 24, 49, 74, 99, np.inf]
labels = [1, 2, 3, 4, 5]
ml_data['runs_class'] = pd.cut(ml_data['Runs'], bins=bins, labels=labels)

# Split the data into features and target variable
X = ml_data.drop(['Runs', 'runs_class'], axis=1)
y = ml_data['runs_class']

# Display the modified DataFrame with runs categorized
ml_data[['Runs', 'runs_class']].head(10), ml_data['runs_class'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

# Splitting the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Show the size of each set
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt

# Define models and their parameters for grid search using ranges
models_params = {
    'Decision Tree': {
        'model': DecisionTreeClassifier(random_state=42),
        'params': {
            'max_depth': [10, 20, 30, None],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4]
        }
    },
    'Random Forest': {
        'model': RandomForestClassifier(random_state=42),
        'params': {
            'n_estimators': [10, 50, 100],
            'max_depth': [10, 20, 30, None],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4]
        }
    },
    'Gradient Boosting': {
        'model': GradientBoostingClassifier(random_state=42),
        'params': {
            'n_estimators': [50, 100, 200],
            'learning_rate': [0.01, 0.1, 0.2],
            'max_depth': [3, 5, 7]
        }
    }
}

# Grid search and results analysis
results = {}
best_params = {}
confusion_matrices = {}

for model_name, mp in models_params.items():
    grid_search = GridSearchCV(mp['model'], mp['params'], cv=5, scoring='accuracy')
    grid_search.fit(X_train, y_train)
    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(X_test)

    # Store best parameters
    best_params[model_name] = grid_search.best_params_

    # Generate and store results
    report = classification_report(y_test, y_pred, output_dict=True)
    results[model_name] = {
        'Macro Precision': f"{report['macro avg']['precision']:.2%}",
        'Macro Recall': f"{report['macro avg']['recall']:.2%}",
        'Macro F1-Score': f"{report['macro avg']['f1-score']:.2%}",
        'Accuracy': f"{report['accuracy']:.2%}"
    }

    # Confusion matrix
    confusion_matrices[model_name] = confusion_matrix(y_test, y_pred)

# Displaying results using Pandas and Seaborn for better visual representation
print("Model Performance Metrics:")
for name, res in results.items():
    print(f"\n{name} results:")
    for key, value in res.items():
        print(f"{key}: {value}")

print("\nBest Parameters:")
for name, params in best_params.items():
    print(f"{name}: {params}")

print("\nConfusion Matrices:")
for name, matrix in confusion_matrices.items():
    plt.figure(figsize=(10,7))
    sns.heatmap(matrix, annot=True, fmt="d", cmap='Blues', linewidths=.5)
    plt.title(f'Confusion Matrix for {name}')
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.show()
```
