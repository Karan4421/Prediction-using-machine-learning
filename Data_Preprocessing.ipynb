```python
df_masterbattingData = pd.concat([df_battingData2023,df_battingData2019,df_battingData2014,df_battingData2010,df_battingData2006,df_battingData2002],ignore_index= True)
```

```python
df_masterbattingData['Dismissal'] = df_masterbattingData['Dismissal'].apply(lambda x: 'o' if pd.notna(x) else 'n')
df_masterbattingData.rename(columns={'Dismissal': 'o/n'}, inplace=True)

df_masterbattingData.head(20)
```

```python
print(df_masterbattingData.head(20))
```

```python
pattern = r'\(c\)|†|\(.*?\)'
df_masterbattingData['Batsman Name'] = df_masterbattingData['Batsman Name'].str.replace(pattern, '', regex=True).str.strip()

# Display the cleaned DataFrame
df_masterbattingData.head(30)
```

```python
total = df_masterbattingData.isnull().sum().sort_values(ascending = False)
percent = (df_masterbattingData.isnull().sum() / df_masterbattingData.isnull().count()).sort_values(ascending = False)
missing_data = pd.concat([total, percent], axis = 1, keys = ['Total', 'Percent'])
missing_data.head(32)
```

```python
# Function to check if the teams in the Match column are in the Team 1 and Team 2 columns
def find_match_id(match):
    teams = match.split(' Vs ')
    match_id = data_mastergameSummary[
        (data_mastergameSummary['Team 1'].isin(teams) & data_mastergameSummary['Team 2'].isin(teams)) |
        (data_mastergameSummary['Team 2'].isin(teams) & data_mastergameSummary['Team 1'].isin(teams))
    ]['id']
    return match_id.iloc[0] if not match_id.empty else None

# Apply function to create a new Match id column in battingData2023
df_masterbattingData['id'] = df_masterbattingData['Match'].apply(find_match_id)

df_masterbattingData.head(20)
```

```python
# Prepare the columns from match summary data for merging
game_summary_columns = data_mastergameSummary[['id', 'Winner', 'VictoryMargin', 'Ground']]

# Merge the batting summary data with the selected match summary columns on 'match_id'
merged_data = pd.merge(df_masterbattingData, game_summary_columns, on='id', how='left')
merged_data.head()
```

```python
total = merged_data.isnull().sum().sort_values(ascending = False)
percent = (merged_data.isnull().sum() / merged_data.isnull().count()).sort_values(ascending = False)
missing_data = pd.concat([total, percent], axis = 1, keys = ['Total', 'Percent'])
missing_data.head(32)
```
